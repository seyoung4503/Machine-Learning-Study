```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('output_gru.csv')

# 'mean_fold'와 'mean_acc' 컬럼 선택
selected_columns = df[['index','mean_fold', 'mean_acc']]

# 선택한 컬럼 데이터 출력
print(selected_columns)
```

```python
import os

dataset_path = './datasets/Classical/'

file_list = os.listdir(dataset_path)
file_list.sort()

file_list = file_list[1:]


data_index = selected_columns['index'].to_numpy()
print(type(data_index[0]))

dataset_name = []
for name in file_list:
    a = name.split('.')
    

    num = a[0].split('_')[0]
    if num in str(data_index):
        dataset_name.append(a[0])

print(data_index)
print(dataset_name)
```
